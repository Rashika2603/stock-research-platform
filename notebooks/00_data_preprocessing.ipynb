# 1. Import Required Libraries

The required Python libraries are imported for data manipulation and preprocessing.

```python
import pandas as pd
import numpy as np
```

# 2. Load Raw Apple Stock Dataset

The raw Apple stock dataset is loaded from the project's data directory. The Date column is parsed as a datetime object and used as the DataFrame index to facilitate time-series analysis.

```python
apple = pd.read_csv(
    "../data/raw/AAPL.csv",
    index_col="Date",
    parse_dates=True
)
```

# 3. Preview the Dataset

The first five observations are displayed to verify that the dataset has been loaded correctly and to gain an initial understanding of its structure.

```python
apple.head()
```

# 4. Dataset Structure

The dataset structure is examined to determine:

- Number of observations
- Number of variables
- Data types
- Missing values
- Memory usage

```python
apple.info()
```

### Key Insights

- The dataset contains **1,628 trading days** spanning from **2 January 2020 to 25 June 2026**.
- Five numerical variables are available: Close, High, Low, Open, and Volume.
- All columns contain complete observations with no missing values.
- The Date column has been successfully stored as a DatetimeIndex, making the dataset suitable for time-series analysis.

# 5. Statistical Summary

Descriptive statistics provide an overview of the central tendency, spread, and overall distribution of each numerical feature.

```python
apple.describe()
```

### Key Insights

# 6. Missing Value Analysis

Missing values are identified to ensure the dataset is complete before analysis.

# 7. Duplicate Record Analysis

Duplicate observations are checked to prevent redundant information from affecting future analyses.

# 10. Date Index Inspection

The Date index is inspected to verify chronological ordering and datetime formatting.

```python
apple.index
```

# 11. Validate Data Types

The data type of each variable is verified.

# 12. Sort Dataset

The dataset is sorted chronologically to ensure the observations follow the correct trading sequence.

```python
apple.sort_index(inplace=True)

apple.head()
```

# 13. Save Clean Dataset

The cleaned dataset is saved to the processed data directory. This dataset will be used throughout the remainder of the project.

```python
apple.reset_index().to_csv(
    "../data/processed/apple_clean.csv",
    index=False
)
```
